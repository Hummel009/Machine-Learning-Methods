1. Найти сумму нечетных цифр введенного натурального числа

```python
def sum_of_not_x2(number):
    total = 0
    for digit in str(number):
        if int(digit) % 2 != 0:
            total += int(digit)
    return total

user_input = input("Enter the num: ")

result = sum_of_not_x2(user_input)

print("Sum of not x2:", result)
```

2.Посчитать, сколько пар (стоят рядом) верхнего и нижнего регистра
находится в веденном с клавиатуры слове. (Пример HjkLM- 1 пара
нижнего, 1 пара верхнего), а также сколько согласных букв в слове.

```python
def count_pairs_and_not_aeiou(word):
    upper_lower_pairs = 0
    lower_upper_pairs = 0
    not_aeiou_count = 0
    aeiou = "aeiouAEIOU"

    for i in range(len(word) - 1):
        if word[i].isupper() and word[i + 1].islower():
            upper_lower_pairs += 1
        elif word[i].islower() and word[i + 1].isupper():
            lower_upper_pairs += 1

    for char in word:
        if char.isalpha() and char not in aeiou:
            not_aeiou_count += 1

    return (upper_lower_pairs, lower_upper_pairs, not_aeiou_count)

user_input = input("Enter the word: ")

upper_lower, lower_upper, not_aeiou = count_pairs_and_not_aeiou(user_input)
print("Upper-to-lower: ", upper_lower)
print("Lower-to-upper: ", lower_upper)
print("Not aeiou: ", not_aeiou)
```

3. Дан список list=[1612,49,'hello',6,19,'world’].
Все числа этого списка проверить на чётность. Если число чётное, то
посчитать сумму его цифр. Если нечётное, то заменить его на 1 в
списке. Все слова: посчитать количество гласных и согласных.
Вывести всё на экран.

```python
def process_list(input_list):
    aeiou = "aeiouAEIOU"
    total_aeiou = 0
    total_not_aeiou = 0

    for i in range(len(input_list)):
        if isinstance(input_list[i], int):
            if input_list[i] % 2 == 0:
               digit_sum = 0
               for digit in str(input_list[i]):
                   digit_sum += int(digit)
               print(f"X2 num {input_list[i]}: sum of digits = {digit_sum}")
            else:
                input_list[i] = 1
                print("Not x2 num replaced with 1")
        elif isinstance(input_list[i], str):
            for char in input_list[i]:
                if char in aeiou:
                    total_aeiou += 1
                elif char.isalpha():
                    total_not_aeiou += 1

    return total_aeiou, total_not_aeiou

input_list = [1612, 49, "hello", 6, 19, "world"]

aeiou_count, not_aeiou_count = process_list(input_list)

print("Renewed list: ", input_list)
print("Aeiou quantity: ", aeiou_count)
print("Not aeiou quantity: ", not_aeiou_count)
```

4. Отсортируйте словарь по значению в порядке возрастания и
убывания.

```python
my_hashmap = {
    'a': 3,
    'b': 1,
    'c': 2,
    'd': 4
}

def sort_rule(item):
    return item[1]

toinf = dict(sorted(my_hashmap.items(), key=sort_rule))
print("From 1 to inf:", toinf)

toone = dict(sorted(my_hashmap.items(), key=sort_rule, reverse=True))
print("From inf to 1:", toone)
```

1. Напишите функцию is_password_good(password), которая принимает
в качестве аргумента строковое значение пароля password и возвращает
значение True если пароль является надежным и False в противном случае.
Пароль является надежным если:
– его длина не менее 10 символов;
– он содержит как минимум одну заглавную букву (верхний регистр);
– он содержит как минимум одну строчную букву (нижний регистр)

```python
def is_password_good(password):
    if len(password) < 10:
        return False

    has_upper = False
    has_lower = False

    for char in password:
        if char.isupper():
            has_upper = True
        if char.islower():
            has_lower = True
        if has_upper and has_lower:
            break

    return has_upper and has_lower

password = input("Enter password: ")
if is_password_good(password):
    print("Password good.")
else:
    print("Password not good.")
```

3. Дан двумерный массив и два числа: i и j. Поменяйте в массиве
столбцы с номерами i и j и выведите результат.
Программа получает на вход размеры массива n и m, затем элементы
массива, затем числа i и j.

```python
def swap_columns(matrix, i, j):
    for row in matrix:
        row[i], row[j] = row[j], row[i]

n = int(input("Enter quantity of lines (n): "))
m = int(input("Enter quantity of columns (m): "))

matrix = []
for _ in range(n):
    row = []
    for _ in range(m):
        item = int(input("Enter a number: "))
        row.append(item)
    matrix.append(row)

i = int(input("Enter index of column 1 (i): "))
j = int(input("Enter index of column 2 (j): "))

if i < 0 or i >= m or j < 0 or j >= m:
    print("Error: index isn't in range of array.")
else:

    print("Matrix before swap:")
    for row in matrix:
        for item in row:
            print(item, end=" ")
        print()

    swap_columns(matrix, i, j)

    print("Matrix after swap:")
    for row in matrix:
        for item in row:
            print(item, end=" ")
        print()
```

4. Напишите программу, демонстрирующую работу try\except\finally

```python
def divide_numbers():
    try:
        num1 = float(input("Enter num 1: "))
        num2 = float(input("Enter num 2: "))

        result = num1 / num2
        print("Result:", result)

    except ZeroDivisionError:
        print("Error, zero division!")

    except ValueError:
        print("Error, not digit!")

    finally:
        print("Always printed.")

divide_numbers()
```

Задание 1:
Создать класс List (список), в котором реализовать методы для работы со
списком (не менее 5).

```python
class List:
    def __init__(self):
        self.items = []

    def append(self, item):
        self.items.append(item)

    def remove(self, item):
        self.items.remove(item)

    def instance(self):
        return self.items

    def clear(self):
        self.items.clear()

    def len(self):
        return len(self.items)

my_list = List()
my_list.append(1)
my_list.append(2)
my_list.append(3)
print(my_list.instance())
my_list.remove(2)
print(my_list.instance())
print(my_list.len())
my_list.clear()
print(my_list.instance())
```

Задание 4:
Придумать класс самостоятельно, реализовать в нем методы экземпляра
класса, статические, методы, методы класса.

```python
class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1

    def instance(self):
        return self.count

    @staticmethod
    def is_positive(number):
        return number > 0

counter = Counter()
counter.increment()
counter.increment()
print(counter.instance())

print(Counter.is_positive(5))
print(Counter.is_positive(-3))
```

Задание 3:
Создать классы: «Книга», «Отдел», «Библиотека». В классах реализовать
следующие методы: добавление, удаление, изменение названия книг из
отделов. Классы должны содержать методы доступа и изменения всех поле.
Предусмотреть метод записи информации в файл.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def change_title(self, new_title):
        self.title = new_title


class Department:
    def __init__(self, name):
        self.name = name
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def delete_book(self, title):
        temp_books = []
        for book in self.books:
            if book.title != title:
                temp_books.append(book)
        self.books = temp_books

    def change_book_title(self, old_title, new_title):
        for book in self.books:
            if book.title == old_title:
                book.change_title(new_title)
                break


class Library:
    def __init__(self):
        self.departments = []

    def add_department(self, department):
        self.departments.append(department)

    def write_to_file(self, filename):
        with open(filename, 'w') as f:
            for dept in self.departments:
                f.write(f"Department: {dept.name}")
                f.write("\n")
                print("Department: ", dept.name)
                for book in dept.books:
                    f.write(f'  - {book.title} by {book.author}\n')
                    print(f'  - {book.title} by {book.author}')
                f.write('\n')
                print()



library = Library()

dept1 = Department("Classic")
dept1.add_book(Book("Voina i mir", "Lev Tolstoi"))
dept1.add_book(Book("Prestuplenie i nakazanie", "Fedor Dostoevskiy"))

dept2 = Department("Non-Classic")
dept2.add_book(Book("Neznaika v solnechnom gorode", "Nosov"))

library.add_department(dept1)
library.add_department(dept2)

library.write_to_file("library_info.txt")

dept1.change_book_title("Voina i mir", "War and peace")

library.write_to_file("library_info.txt")
```
